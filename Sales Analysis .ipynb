```python
# month and year from  Data Set 
df['month_year'] = df ['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df.groupby('month_year').sum(numeric_only=True)
```

```python
#Grouping Month & year
df_Trend = df.groupby('month_year')['sales'].sum()
```

```python
df_Trend['month_year'] = pd.to_datetime(df_Trend['month_year'])

# Sort values to maintain chronological order
df_Trend = df_Trend.sort_values('month_year')

# Plot the trend
plt.figure(figsize=(10,5))  # Optional: Adjusts figure size
plt.plot(df_Trend['month_year'], df_Trend['sales'], marker='o', linestyle='-')

# Formatting the plot
plt.xlabel('Month-Year')
plt.ylabel('Sales')
plt.title('Sales Trend Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show the plot
plt.show()
```

```python
most_profit_cat = df.groupby(['category', 'sub_category']).sum(numeric_only=True)[['profit']]
most_profit_cat = most_profit_cat.sort_values('profit', ascending=False)

# Reset index for visualization
most_profit_cat = most_profit_cat.reset_index()

# Select top 10 most profitable sub-categories
top_10 = most_profit_cat.head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x='profit', y='sub_category', hue='category', data=top_10, palette='magma')

plt.xlabel("Total Profit")
plt.ylabel("Sub-Category")
plt.title("Top 10 Most Profitable Sub-Categories")
plt.show()
```
